```python
Cross_schema = (StructType()
.add('Facility ID', StringType(), True)
.add('State', StringType(), True)
.add('County Name', StringType(), True)
.add('FIPS', StringType(), True))
```

```python
Cross = spark.read.csv('CMS_Facility_to_FIPS.csv', header=True, schema=Cross_schema)
Cross.take(1) >> to_pandas
```

```python
@pipeable
def spread(val_col, var_col, group_by_col, df):
    return (df
            .groupBy(group_by_col)
            .pivot(val_col)
            .sum(var_col))
```

```python
Poverty_Cleaned = (Poverty
                    .select('FIPS_code', 'PCTPOVALL_2020')
                    .withColumnRenamed('FIPS_code', 'FIPS'))
Poverty_Cleaned.collect() >> to_pandas
```

```python
Measures = ['Average (median) time patients spent in the emergency department before leaving from the visit A lower number of minutes is better', 'Left before being seen']

Timley_Hospital = (((Hospital
    .select('Facility ID', 'County Name', 'Condition', 'Measure Name', 'Score', 'State')
    .where(col('Condition') == 'Emergency Department')
    .where(col('Measure Name').isin(Measures))
    .drop(col('Condition')))
    
>> spread('Measure Name', 'Score', ['Facility ID', 'County Name', 'State']))
.withColumnRenamed('Average (median) time patients spent in the emergency department before leaving from the visit A lower number of minutes is better', 'Avg_Time_Spent')
.join(Cross, on=['Facility ID'], how='left')
.join(Poverty_Cleaned, on='FIPS', how='left'))

Timley_Hospital.collect() >> to_pandas
```
